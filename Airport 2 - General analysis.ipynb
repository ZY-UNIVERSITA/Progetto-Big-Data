### **3. Airports**

**Number of airports**

```python
# Find all airports, combining ORIGIN and DESTINATION IATA code.
# Without ignore_index, only half of rows would be shown.
aeroporti = pd.concat([df['ORIGIN_IATA'], df['DEST_IATA']], ignore_index = True)

aeroporti
```

```python
# Unique airports
aeroporti.nunique()
```

**Number of flight per airports**

```python
# Number of departure flight
partenze_count = df['ORIGIN_IATA'].value_counts()

# Number of arriving flight
arrivi_count = df['DEST_IATA'].value_counts()

# Combining the 2 series into 1, using fill_value = 0, each airport is only counted once
numero_voli_per_aeroporto = partenze_count.add(arrivi_count, fill_value=0)

numero_voli_per_aeroporto
```

**Visualize the 10 busiest airports using a plot**

```python
# Bar plot
numero_voli_per_aeroporto.nlargest(10).plot(kind='bar', legend=False, color='skyblue')

# Add label and title
plt.title('Numero di voli per i principali 10 aeroporti USA')
plt.xlabel('Codice IATA dell\'aeroporto')
plt.ylabel('Numero di voli')
plt.xticks(rotation=0)

# Show graph
plt.show()
```

**Busiest airport per departure flights**

```python
#  Busiest airport per departure
partenze_count.nlargest(10).plot(kind='bar', legend=False, color='skyblue')

plt.title('Numero di voli in entrata per i principali 10 aeroporti USA')
plt.xlabel('Codice IATA dell\'aeroporto')
plt.ylabel('Numero di voli')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

**Busiest airport per arriving flights**

```python
# Busiest airports per arriving
arrivi_count.nlargest(10).plot(kind='bar', legend=False, color='skyblue')

plt.title('Numero di voli in uscita per i principali 10 aeroporti USA')
plt.xlabel('Codice IATA dell\'aeroporto')
plt.ylabel('Numero di voli')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

**Comparing departure and arriving for the 10 busiest airports**

```python
# Create a new df for arrivi and partenze
dataPartenzeArrivi = pd.DataFrame({
    'Arrivi': arrivi_count.nlargest(10),
    'Partenze': partenze_count.nlargest(10)
})

# Plot a bar using the 2 series
dataPartenzeArrivi.plot(kind='bar', figsize=(15, 6), title="Partenze e arrivi")
plt.ylabel('Numero di partenze e arrivi')
plt.xlabel('Aeroporto')
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

#### **4. Busiest time**

**Busiest day of the week**

```python
# Create a new column for day of week for each flight
# 0: Monday and 6: Sunday
df['DAY_OF_WEEK'] = df['PROGRAMMED_DEPARTING_TIME_UTC'].dt.dayofweek
```

```python
# Group flight per day of week
voli_per_giorno_della_settimana = df.groupby('DAY_OF_WEEK')['FLIGHTS'].count()
```

```python
# Bar chart to visualize flights by day of week
voli_per_giorno_della_settimana.plot(kind='line')

plt.title('Volume di voli per giorno della settimana')
plt.xlabel('Giorno della settimana')
plt.ylabel('Numero di voli')
plt.xticks(rotation=0)

plt.show()
```

```python
# Busiest day
voli_per_giorno_della_settimana.sort_values(ascending=False)
```

**Busiest hour of the week**

```python
# Create a new column for day of week for each flight
df["HOUR_OF_DAY"] = df['PROGRAMMED_DEPARTING_TIME_UTC'].dt.hour
```

```python
# Group by hour of day
voli_per_ora_del_giorno = df.groupby('HOUR_OF_DAY')['FLIGHTS'].count()
```

```python
# Bar chart to visualize flights by hour of day
voli_per_ora_del_giorno.plot(kind='bar')

plt.title('Volume di voli per giorno della settimana')
plt.xlabel('Giorno della settimana')
plt.ylabel('Numero di voli')
plt.xticks(rotation=0)

plt.show()
```

```python
voli_per_ora_del_giorno.sort_values(ascending=False)
```

**Busiest hour of day and day of week using heatmap**

```python
voli_per_ora_e_per_giorno = df.groupby(['DAY_OF_WEEK', 'HOUR_OF_DAY'])['FLIGHTS'].count().unstack()
```

```python
voli_per_ora_e_per_giorno
```

```python
# Heatmap for flights by hour and day of week
plt.figure(figsize=(12, 6))
sns.heatmap(voli_per_ora_e_per_giorno, cmap='viridis_r')
plt.title('Attività per giorno della settimana e per ora del giorno')
plt.xlabel('Ora del giornno')
plt.ylabel('Giorno della settimana')
plt.show()
```

#### **5. Most popular routes**

```python
# Create routes and count flights per route
group_columns = ['ORIGIN_IATA', 'ORIGIN_CITY_NAME', 'DEST_IATA', 'DEST_CITY_NAME' ]
route_counts = df.groupby(group_columns)['FLIGHTS'].sum().reset_index()

# Ordering routes by flights numbers
busiest_routes = route_counts.sort_values(by='FLIGHTS', ascending=False)

# Show busiest routes
busiest_routes.head(10)
```

```python
# Plot a bar to visualize busisest routes
top_routes = busiest_routes.head(10)

# Crea il plot
plt.figure(figsize=(12, 8))
plt.barh(top_routes['ORIGIN_CITY_NAME'] + ' to ' + top_routes['DEST_CITY_NAME'], top_routes['FLIGHTS'], color='skyblue')

# Configuration fo bar plot
plt.title('10 rotte più trafficate')
plt.xlabel('Numero di voli')
plt.ylabel('Rotte')
plt.gca().invert_yaxis()
plt.show()

```

#### **6. Distribution of flights per states**

The state boundary shapefiles used in this analysis were sourced from the official U.S. Census Bureau: <a href="https://www.census.gov/cgi-bin/geo/shapefiles/index.php?year=2024&layergroup=States+%28and+equivalent%29 ">website</a>.

**Flights per states**

```python
# Number of departures per states
stati_partenze_count = df['ORIGIN_STATE_NM'].value_counts()

# Number of arriving per states
stati_arrivi_count = df['DEST_STATE_NM'].value_counts()

# Total number of flights per states. Convert series to dataframe
numero_voli_per_stati_df = pd.DataFrame(stati_partenze_count.add(stati_arrivi_count, fill_value=0))

numero_voli_per_stati_df
```

**Importazione dei file per costrutire la forma dei singoli stati**

```python
# Without colab
# gdf = gpd.read_file(filename=r"/shapefiles/tl_2024_us_state.shp")
```

```python
# Using colab

# Read USA shapefiles
gdf = gpd.read_file(filename=r"/content/drive/MyDrive/shapefiles/tl_2024_us_state.shp")
```

Removing not significative island states

```python
gdf2 = gdf.loc[ (gdf["NAME"] != "Guam") &
                (gdf["NAME"] != "United States Virgin Islands") &
                (gdf["NAME"] != "American Samoa") &
                (gdf["NAME"] != "Commonwealth of the Northern Mariana Islands") &
                (gdf["NAME"] != "Alaska")]
```

```python
gdf_region = gpd.GeoDataFrame(gdf2, geometry="geometry")
gdf_region.plot()
```

```python
# Merge states with shapesfiles
df_merge = numero_voli_per_stati_df.merge(right=gdf_region, left_on="ORIGIN_STATE_NM", right_on="NAME", how="inner")
gdf_merge = gpd.GeoDataFrame(df_merge, geometry='geometry')
```

```python
# Add labels
fig, ax = plt.subplots(1, 1, figsize=(20, 10), dpi=300)
gdf_merge.plot(column='count', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add labels for each region
for x, y, label in zip(gdf_merge.geometry.centroid.x, gdf_merge.geometry.centroid.y, gdf_merge['NAME']):
    ax.text(x, y, label, fontsize=8, ha='center', color='black')

plt.title("Numero di voli per stati")
plt.show()
```
